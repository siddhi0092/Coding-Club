```python
#queue using 2 stacks

class TwoStackQueue:
    def __init__(self):
        self.stack1 = []
        self.stack2 = []
    
    def is_empty(self):
        return len(self.stack1) == 0 and len(self.stack2) == 0
    
    def enqueue(self, item):
        self.stack1.append(item)
    
    def dequeue(self):
        if self.is_empty():
            print("Empty!!")
            return
        if len(self.stack2) == 0:    
            while self.stack1:
                self.stack2.append(self.stack1.pop())
        return self.stack2.pop()
```
