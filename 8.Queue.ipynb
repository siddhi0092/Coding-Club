```python
class Queue:
    def __init__(self,size):
        self.size = size
        self.queue = []  
    
    def is_empty(self):
        return len(self.queue) == 0
    
    def is_full(self):
        return len(self.queue) == self.size
    
    def enqueue(self, data):
        self.queue.append(data)
        return f"{data} appended"
    
    def dequeue(self):
        item = self.queue.pop(0)
        return f"{item} popped"
    
    def display(self):
        print(self.queue)
        print("Rear : ",self.queue[-1])
        print("Front : ", self.queue[0])

h1 = Queue(5)
# h1.display()  #gives out of range before we perform any operation
h1.enqueue(3)
h1.display()
#h1.append(3)
```

```python
#dequeue

from collections import deque

class CircularQueue:
    def __init__(self,size):
        self.queue = deque(maxlen = size)
        self.size = size
    
    def is_empty(self):
        return len(self.queue) == 0
        # return "Queue is not empty"   giving error cause it is always true 
    
    def is_full(self):
        return len(self.queue) == self.size
    
    def enqueue(self, data):
        if self.is_full():
            return self.is_full()
        self.queue.append(data)
    
    def dequeue(self):
        if self.is_empty():
            return self.is_empty()
        self.queue.popleft()
    
    def display(self):
        if self.is_empty():
            return self.is_empty()
        print(self.queue)
        print("Rear:", self.queue[-1])
        print("Front:", self.queue[0])
    
h = CircularQueue(5)
h.enqueue(10)
h.enqueue(20)
h.enqueue(30)
h.display()
```
