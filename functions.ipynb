```python
#  variable length arguments (tuple)

#  +,-,/,//,%,*

def add(a,b):
    return a+b
def sub(a,b):
    return a-b
def div(a,b):
    return a/b
def floordivison(a,b):
    return a//b
def modulus(a,b):
    return a%b
def mul(*a):
    result = 1
    for i in a:
        result = result * i
    return result

operator = input("Enter from the following operators: +, -, /, //, %, *")
a = input("Enter a")
b = input("Enter b")

if operator == '+':
    print(add(int(a), int(b)))
elif operator == '-':
    print(sub(int(a), int(b)))
elif operator == '/':
    print(div(int(a), int(b)))
elif operator == '//':
    print(floordivison(int(a), int(b)))
elif operator == '%':
    print(modulus(int(a), int(b)))
elif operator == '*':
    print(mul(int(a), int(b)))
else:
    print("Invalid operator")
```
