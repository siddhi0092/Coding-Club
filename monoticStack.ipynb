```python
#Next Smallest Monotonic Stack

arr = [4,5,2,10,8]
s = []
result = [-1]*len(arr)

for i in range(len(arr)):
    # we dont know the range therefore use while
    while s and arr[i] < arr[s[-1]]:
        index = s.pop()
        result[index] = arr[i]
    s.append(i)

print(result)
```

```python
#Next Greastest Monotonic Stack

arr = [4,5,2,10,8]
s = []
result = [-1]*len(arr)

for i in range(len(arr)):
    # we dont know the range therefore use while
    while s and arr[i] > arr[s[-1]]:
        index = s.pop()
        result[index] = arr[i]
    s.append(i)

print(result)
```

```python
#Smallest and greatest using monotonic stack
class Monotonic:
    def __init__(self):
        self.stack = []
        self.result = []
    
    def next_smallest(self, nums):
        self.stack = []
        self.result = [-1]*len(nums)
        for i in range(len(nums)):
            while self.stack and nums[i] < nums[self.stack[-1]]:
                index = self.stack.pop()
                self.result[index] = nums[i]
            self.stack.append(i)
        return self.result
    
    def next_greatest(self, nums):
        self.stack = []
        self.result = [-1]*len(nums)
        for i in range(len(nums)):
            while self.stack and nums[i] > nums[self.stack[-1]]:
                index = self.stack.pop()
                self.result[index] = nums[i]
            self.stack.append(i)
        return self.result

s = Monotonic()
nums = [4,5,2,10,8]
print(s.next_smallest(nums))
print(s.next_greatest(nums))
        
```
