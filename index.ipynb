```python
#Rock paper and scissors 

n = int(input("Enter how many times you want to play the game"))
win1 = 0
win2 = 0
for i in range(n):
    p1 = input("P1 : R,P,S")
    p2 = input("P2 : R,P,S")

    if p1 == "R" and p2 == "P":
        print("p2 wins")
        win2 += 1
    elif p1 == "P" and p2 == "R":
        print("p1 wins")
        win1 += 1
    elif p1 == "P" and p2 == "S":
        print("p2 wins")
        win2 += 1
    elif p1 == "S" and p2 == "P":
        print("p1 wins")
        win1 += 1
    elif p1 == "S" and p2 == "R":
        print("p2 wins")
        win2 += 1
    elif p1 == "R" and p2 == "S":
        print("p1 wins")
        win1 += 1
    elif p1 == p2:
        print("tie!")
    else:
        print("Choose Valid")

print("score of p1 =", win1)
print("score of p2 =", win2)

if max(win1,win2) == win1:
    print("P1 is the winner")
else:
    print("p2 is the winner")
```
