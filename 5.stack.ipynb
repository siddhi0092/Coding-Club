```python
# stack using functions 

def push(stack, item):
    stack.append(item)
    print("Pushed:", item)

def pop(stack):
    if isEmpty(stack):
        print("Stack is empty")
    else:
        a = stack.pop()
        print("Popped:", a)

def peek(stack):
    if isEmpty(stack):
        print("Stack is empty")
    else:
        print("Top element is:", stack[-1])

def isEmpty(stack):
    return len(stack) == 0

def size(stack):
    return len(stack)

def display(stack):
    if isEmpty(stack):
        print("Stack is empty")
    else:
        for i in reversed(stack):
            print(i)

stack = []

while True:
    print("\nChoose from the following operations")
    print("1 - Push")
    print("2 - Pop")
    print("3 - Peek")
    print("4 - Is Empty")
    print("5 - Size")
    print("6 - Display")
    print("7 - Exit")

    a = int(input("Enter your choice: "))

    if a == 1:
        item = int(input("Enter item to push inside stack: "))
        push(stack, item)

    elif a == 2:
        pop(stack)

    elif a == 3:
        peek(stack)

    elif a == 4:
        if isEmpty(stack):
            print("Stack is empty")
        else:
            print("Stack is not empty")

    elif a == 5:
        print("Size of stack is:", size(stack))

    elif a == 6:
        print("Stack elements are:")
        display(stack)

    elif a == 7:
        print("Exiting program...")
        break

    else:
        print("Not valid choice")
```

```python
#using class

class Stack:
    
    def __init__(self):
        self.items = []  #initializing the stack
    
    def push(self, item):
        self.items.append(item)
        print("Pushed")
    
    def pop(self):
        if self.is_empty():    #to include any method in another method, use self.
            print("Stack is empty")
        else:
            return self.items.pop()
    
    def is_empty(self):
        return len(self.items) == 0
    
    def peek(self):
        if self.is_empty():
            print("Stack is empty")
        else:
            return self.items[-1]
    
    def size(self):
        return len(self.items)
    
    def display(self):
        if self.is_empty():
            print("Stack is empty")
        else:
            print("ele from top to bottom")
            for i in reversed(self.items):
                print(i)
    

st = Stack()
st.push(10)
st.push(20)
st.push(30)
st.pop()
st.pop()
st.pop()
st.peek()
print(st.is_empty())
```

```python
# Stack using Fixed Size

SIZE = 5
stack = [None] * SIZE
top = -1


def push(item):
    global top  
    if top == SIZE - 1:
        print("Stack Overflow")
    else:
        top += 1
        stack[top] = item
        print("Pushed:", item)

def pop():
    global top
    if is_empty():
        print("Stack Underflow")
    else:
        popped_item = stack[top]
        stack[top] = None
        top -= 1
        print("Popped:", popped_item)


def is_empty():
    return top == -1


def is_full():
    return top == SIZE - 1


def peek():
    if is_empty():
        print("Stack is empty, nothing to peek")
    else:
        print("Top element:", stack[top])


def display():
    if is_empty():
        print("Stack is empty")
    else:
        print("Stack elements:")
        for i in range(top, -1, -1):
            print(stack[i])
```

```python
#Fixed stack

class FixedStack:
    def __init__(self, size):  #used to initialize all the attributes 
        self.size = size
        self.stack = [None] * size
        self.top = -1
    
    def push(self, item):
        if self.isFull():
            return "Stack is full"
        else:
            self.top += 1
            self.stack[self.top] = item
            return f"{item} pushed"
    
    def isFull(self):
        if self.top == self.size - 1:
            return True
        else:
            return False

    def pop(self):
        if self.isEmpty():
            return "Stack is empty"
        else:
            # stack.top -= 1
            # self.stack[self.top] = None
            # print("Item Popped") 
            item = self.stack[self.top]
            self.top -= 1
            print(f"{item} popped")

    def isEmpty(self):
        if self.top == -1:
            return True
        else:
            return False
        
    def peek(self):
        if self.isEmpty():
            return "Stack is empty"
        else:
            return self.stack[self.top]
    
    def display(self):
        if self.isEmpty():
            print("Stack is empty")
        else:
            for i in range(self.top, -1, -1):
                print(f"{self.stack[i]}", end = ", ")

s = FixedStack(10)
print(s.isEmpty())
print(s.peek())
print(s.push(20)) 
print(s.push(30)) 
print(s.push(40)) 
print(s.push(50)) 
print(s.push(60)) 
print(s.push(70)) 
print(s.push(80))
print(s.push(90)) 
print(s.push(100)) 
print(s.push(120)) 
print(s.push(130)) 
s.display()
```
