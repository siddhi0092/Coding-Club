```python
#Single Linked List

class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
        # self.next = False
        # self.next = 0       #we can use these three ways as well
        # self.next = []
    
class SingleLinkedList:
    def __init__(self):
        self.head = None
    
    #INSERTION

    def insertBegin(self,data):   #insert the data at the beginning
        newNode = Node(data)     #create a new node
        newNode.next = self.head   #point the newNode next to prev head
        self.head = newNode       #change the head to newNode

    def insertEnd(self,data):     #insert the data at the end
        newNode = Node(data)      
        temp = self.head
        while temp.next:          #traverse to the last node
            temp = temp.next
        temp.next = newNode       #point the last node next to newNode
    
    def insert_position(self,positon,data):
        newNode = Node(data)     
        temp = self.head
        for i in range(positon-1): #traverse to the position-1 node
            temp = temp.next
        newNode.data = data    #assign data to newNode
        newNode.next = temp.next    #point newNode next to position-1 node next
        temp.next = newNode        #point position-1 node next to newNode

    #DISPLAY

    def display(self):
        temp = self.head
        while temp:
            print(temp.data, end=" --> ")
            temp = temp.next
        print("None")

    #DELETION

    def deleteBegin(self):
        temp = self.head
        self.head = temp.next   #point head to the next node of current head
        temp.next = None

    def deleteEnd(self):    #you can you prev and temp
        prev = self.head
        temp = self.head.next 
        while temp.next is not None:   #traverse to the last node            
            temp = temp.next
            prev = temp.next
        prev.next = None   
    
    def delete_position(self,position):   #take prev and next
        prev = self.head
        temp = self.head.next
        for i in range(position-1): #traverse to the position-1 node
            temp = temp.next
            prev = prev.next
        prev.next = temp.next    #point position-1 node next to position+1 node

    #reverse single linked list 
    def reverse_list(self):
        prev = None
        current = self.head
        while current:
            next_node = current.next  #store the next node
            current.next = prev       #reverse the link
            prev = current            #move prev to current
            current = next_node       #move current to next node
        self.head = prev             #update head to the new first node

sll = SingleLinkedList() 

node = Node(80)
sll.head = node
n1 = Node(90)
node.next = n1
n2 = Node(70)
n1.next = n2
n3 = Node(60)
n2.next = n3


sll.insertBegin(10)
# sll.insertBegin(20)
sll.insertEnd(30)
sll.insert_position(1,25)
sll.display()
```

```python
#double linked list

class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None

    #INSERTION

    def insert_at_beginning(self, data):
        new_node = Node(data)
        
        if self.head is not None:
            self.head.prev = new_node
            new_node.next = self.head
        
        self.head = new_node
    
    def insert_at_end(self, data):
        new_node = Node(data)

        if self.head is None:
            self.head = new_node
            return
        
        temp = self.head
        while temp.next:
            temp = temp.next
        
        temp.next = new_node
        new_node.prev = temp
    
    def insert_at_position(self, pos, data):
        if pos <= 1:
            self.insert_at_beginning(data)
            return

        new_node = Node(data)
        temp = self.head
        count = 1

        while temp and count < pos - 1:
            temp = temp.next
            count += 1

        if temp is None:
            print("Position out of range")
            return

        new_node.next = temp.next
        new_node.prev = temp

        if temp.next:
            temp.next.prev = new_node

        temp.next = new_node

    #DELETION

    def delete_from_beginning(self):
            if self.head is None:
                print("List is empty")
                return

            self.head = self.head.next
            if self.head:
                self.head.prev = None

    def delete_from_end(self):
        if self.head is None:
            print("List is empty")
            return

        temp = self.head
        while temp.next:
            temp = temp.next

        if temp.prev:
            temp.prev.next = None
        else:
            self.head = None
    
    def delete_by_value(self, key):
        temp = self.head

        while temp:
            if temp.data == key:
                if temp.prev is None:  # head node
                    self.delete_from_beginning()
                elif temp.next is None:  # last node
                    self.delete_from_end()
                else:
                    temp.prev.next = temp.next
                    temp.next.prev = temp.prev
                return
            temp = temp.next

        print("Value not found")

        if temp.next is None:
            self.head = None
            return

        while temp.next:
            temp = temp.next

        temp.prev.next = None

    def delete_by_value(self, key):
        temp = self.head

        while temp:
            if temp.data == key:
                if temp.prev is None:  # head node
                    self.delete_from_beginning()
                elif temp.next is None:  # last node
                    self.delete_from_end()
                else:
                    temp.prev.next = temp.next
                    temp.next.prev = temp.prev
                return
            temp = temp.next

        print("Value not found")

    # DISPLAY

    def display(self):
        temp = self.head
        if temp is None:
            print("List is empty")
            return

        while temp:
            print(temp.data, end=" <-> ")
            last = temp
            temp = temp.next
        print("None")

    
    #reversing a double linked list 
    def reversing_double_linked_list(self):
        current = self.head
        temp = None

        while current is not None:
            temp = current.prev
            current.prev = current.next
            current.next = temp
            current = current.prev

        if temp is not None:
            self.head = temp.prev
    
    #finding the middle element 
    def find_middle(self):
        slow = self.head
        fast = self.head

        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
        
        return slow.data


dll = DoublyLinkedList()
dll.insert_at_end(10)
dll.insert_at_end(20)
dll.insert_at_end(30)
dll.insert_at_beginning(5)
dll.insert_at_position(2, 15)
dll.display()
dll.reversing_double_linked_list()
dll.display()
dll.find_middle()
```
