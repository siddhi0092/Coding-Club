PATTERNS

```python
n = 5

for i in range(n):
    for j in range(n-i,0,-1): # i = 0, j = 5,4,3,2,1, i = 1, j = 4,3,2,1
        print(" ", end="")
    for j in range(2*i+1,0,-1):
        print("*", end="")
    print()

for i in range(n):
    for j in range(n//2+1):
        print(" ", end="")
    for j in range(n):
        print("*",end="")
    print()
```

```python
#list comprehension

#square of the given numbers 

given = [1,2,3,4,5,6,7,8]

print([x*x for x in given])
```

```python
# list comprehension (even numbers only)

given = [1,2,3,4,5,6,7,8]

print([x for x in given if x%2 == 0])
```
