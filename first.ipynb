```python
#bubble sorting
#algo 
#check if i > i+1, if yes then swap  

a = [5,4,3,2,1,6,10]
max = 0
min = 0
#ascending order
for i in range(len(a)):
    for j in range(0, len(a)-i-1):
        if a[j] > a[j+1]:
            a[j], a[j+1] = a[j+1], a[j]
print(a)

#descending order
for i in range(len(a)):
    for j in range(0, len(a)-i-1):
        if a[j] < a[j+1]:
            a[j], a[j+1] = a[j+1], a[j]
print(a)
```
