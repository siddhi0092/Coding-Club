```python
# stack using functions 

def push(stack, item):
    stack.append(item)
    print("Pushed:", item)

def pop(stack):
    if isEmpty(stack):
        print("Stack is empty")
    else:
        a = stack.pop()
        print("Popped:", a)

def peek(stack):
    if isEmpty(stack):
        print("Stack is empty")
    else:
        print("Top element is:", stack[-1])

def isEmpty(stack):
    return len(stack) == 0

def size(stack):
    return len(stack)

def display(stack):
    if isEmpty(stack):
        print("Stack is empty")
    else:
        for i in reversed(stack):
            print(i)

stack = []

while True:
    print("\nChoose from the following operations")
    print("1 - Push")
    print("2 - Pop")
    print("3 - Peek")
    print("4 - Is Empty")
    print("5 - Size")
    print("6 - Display")
    print("7 - Exit")

    a = int(input("Enter your choice: "))

    if a == 1:
        item = int(input("Enter item to push inside stack: "))
        push(stack, item)

    elif a == 2:
        pop(stack)

    elif a == 3:
        peek(stack)

    elif a == 4:
        if isEmpty(stack):
            print("Stack is empty")
        else:
            print("Stack is not empty")

    elif a == 5:
        print("Size of stack is:", size(stack))

    elif a == 6:
        print("Stack elements are:")
        display(stack)

    elif a == 7:
        print("Exiting program...")
        break

    else:
        print("Not valid choice")
```

```python
#using class

class Stack:
    
    def __init__(self):
        self.items = []  #initializing the stack
    
    def push(self, item):
        self.items.append(item)
        print("Pushed")
    
    def pop(self):
        if self.is_empty():    #to include any method in another method, use self.
            print("Stack is empty")
        else:
            return self.items.pop()
    
    def is_empty(self):
        return len(self.items) == 0
    
    def peek(self):
        if self.is_empty():
            print("Stack is empty")
        else:
            return self.items[-1]
    
    def size(self):
        return len(self.items)
    
    def display(self):
        if self.is_empty():
            print("Stack is empty")
        else:
            print("ele from top to bottom")
            for i in reversed(self.items):
                print(i)
    

st = Stack()
st.push(10)
st.push(20)
st.push(30)
st.pop()
st.pop()
st.pop()
st.peek()
print(st.is_empty())
```
