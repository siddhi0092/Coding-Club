```python
#Single level inheritance
class parentclass:
    def h1(self):
        print("this is a parent class")

class childclass(parentclass):
    def h2(self):
        print("This is a child class")

b1 = childclass()
b1.h1()
b1.h2()
```

```python
#Multiple Inheritance : child inherits attributes from more than one parent class

class parent1:
    def h1(self):
        print("This is parent 1")

class parent2:
    def h2(self):
        print("This is parent 2")

class child(parent1, parent2):
    def h3(self):
        print("This is child")

ch = child()
ch.h3()
ch.h2()
ch.h1()
```

```python
#Multi-level Inheritance : child inherits from parent, parent inherits from grandparent

class grandparent():
    def h1(self):
        print("Hi, I am grandparent")

class parent(grandparent):
    def h2(self):
        print("Hi, I am Parent")

class child(parent):
    def h3(self):
        print("Hi, I am child")

son = child()
son.h1()
son.h2()
son.h3()
```

```python
# We have overriding in python 
# overloading - same method, different parameters 

#overloading example using default arguments
def demo(a,b=10):
    print(a+b)
demo(12)
demo(12,13)
```

```python
class demo:
    def d1(a,b=10):
        print(a+b)
demo.d1(10)
demo.d1(10,5)
```

```python
#overloading example using variable length arguments
def demo(*a):
    for i in a:
        print(i)
    print("DONE")
demo(10)
demo(10,21)
demo(10,12,45,7)
```

```python
#overloading inside a class  
class demo:
    def dmo(*a):
        for i in a:
            print(i)
        print("DONE")
demo.dmo(10)
demo.dmo(10,21)
```

```python
#Method Overriding
class Person:
    def __init__(self, name):
        self.name = name
    
    def h1(self):
        print("Name : ",self.name)

class Student(Person):
    def __init__(self,name,roll):
        self.name = name
        self.roll = roll
    
    def h1(self):   #we can rewrite the method name
        print("Welcome", self.name)

d = Student("Siddhi",18)
d.h1()
```

```python
#polymorphism --> no. of classes = no. of objects

class Tomato:
    def color(self):
        print("I am Red")
    def ripe(self):
        print("I am rip")

class Onion:
    def color(self):
        print("I am Purple")
    def ripe(self):
        print("I am not rip")

class Banana:
    def color(self):
        print("I am yellow")
    def ripe(self):
        print("I am banana and i am rip")

# tom = Tomato()
# on = Onion()
# ban = Banana()

obj = [Tomato(), Onion(), Banana()]
for i in obj:
    i.color()
    i.ripe()

```
