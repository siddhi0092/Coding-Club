```python
#using modulo

def hash_table(key,size):
    return key % size

size = 10
hash_list = [None]*size
value = "SRI"
key = 27
index = hash_table(key,size)
hash_list[index] = value
print(hash_list)
```

```python
#multiplication
#using floor

import math 

def hash_table(key, size):
    a = 0.618   #common constant 
    return math.floor(size*((key*a)%1))

print(hash_table(29,10))
```

```python
#instead of floor using int 

def hash_table(key,size):
    a = 0.618
    return int(size*((key*a)%1))

print(hash_table(29,10))
```

```python
#collision hashing  -> closed addressing (opened hashing) and open addressing (closed hashing)

#closed addressing:
# concept of chaining --> each and every column is going to store one more list list inside list 
# [[[],[],[]],[],[],[]]  -> 2D list  #separate chaining 


#open addressing *(closed hashing):


#linear probing:
#if no collision then use ->   h(k) = k % size
#--> if collision then use               h(k) = (k+1)%size

#quadric probing:
#if no collision then use ->   h(k) = k % size
#if collision then use -> h(k) = (k + i^2) % size

#double hashing
#if no collision then use ->   h(k) = k % size
#(h1(k) + h2(k)) % size    --> h1(k) = k % size, h2(k) = (k+1)% size or h2(k) = k % (size-1)
# 4 ways
```

```python
#closed addressing

size =  5
hash_table = [[] for _ in range(size)]

def inserting(key):
    index = key % size
    hash_table[index].append(key)

inserting(3)
inserting(4)
inserting(5)
print(hash_table)
```

```python
#linear probing 

size = 5

hash_list = [None]*size
def inserting(key):
    index = key % size
    while hash_list[index] is not None:
        index = (index + 1) % size
    
    hash_list[index] = key

inserting(3)
inserting(4)
inserting(5)
print(hash_list)
```

```python
#quadratic probing 

size = 10
hash_list = [None]*size

def inserting(key):
    index = key % size
    i = 1
    while hash_list[index] is not None:
        # index = (key % size + i*i)% size
        index = (index+i*i)%size
        i += 1
    hash_list[index] = key

def insertingString(key,value):
    index = key % size
    i = 1
    while hash_list[index] is not None:
        # index = (key % size + i*i)% size
        index = (index+i*i)%size
        i += 1
    hash_list[index] = value


insertingString(27, "Siddhi")
inserting(3)
inserting(4)
inserting(5)
inserting(5)
inserting(5)

print(hash_list)
```

```python
#quadratic probing 
#function inside function 

size = 10
hash_list = [None]*size

def inserting(key,value):
    def hash_table(k):
        return k%size
    
    index = hash_table(key)
    i = 1
    while hash_list[index] is not None:
        # index = (key % size + i*i)% size
        index = (index+i*i)%size
        i += 1
    hash_list[index] = value

def insertingString(key,value):
    index = key % size
    i = 1
    while hash_list[index] is not None:
        # index = (key % size + i*i)% size
        index = (index+i*i)%size
        i += 1
    hash_list[index] = value

insertingString(27, "Siddhi")
inserting(5, "siddhi")
inserting(3, "siddhihehe")

print(hash_list)
```
