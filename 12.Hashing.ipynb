```python
# print(hash("sri"))

#direct hashing

hash_table = [None]*10
key = 4
hash_table[key] = "Coding Club"
print(hash_table)
print()
hash_table[key-1] = "Siddhi"
hash_table[key-2] = "Siddhi"
hash_table[key-3] = "Siddhi"
hash_table[key+1] = "Siddhi"

print(hash_table)
```

```python
#using modulo

def hash_table(key,size):
    return key % size

size = 10
hash_list = [None]*size
value = "SRI"
key = 27
index = hash_table(key,size)
hash_list[index] = value
print(hash_list)
```

```python
#multiplication
#using floor

import math 

def hash_table(key, size):
    a = 0.618   #common constant 
    return math.floor(size*((key*a)%1))

print(hash_table(29,10))
```

```python
#instead of floor using int 

def hash_table(key,size):
    a = 0.618
    return int(size*((key*a)%1))

print(hash_table(29,10))
```
